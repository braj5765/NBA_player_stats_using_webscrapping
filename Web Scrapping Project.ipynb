# GENERAL METHOD

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2019_per_game.html'
```

### Read HTML webpage into pandas


```python
import pandas as pd
```

```python
df = pd.read_html(url, header = 0)
df
```

```python
len(df)
```

```python
df2019=df[0]
df2019
```

### Data Cleaning

```python
## We are removing the multiple table column headings present inbetween the data
df2019[df2019.Age == 'Age']
```

```python
len(df2019[df2019.Age == 'Age'])
```

```python
## we will use the drop function to remove these 26 rows
df = df2019.drop(df2019[df2019.Age == 'Age'].index)
```

```python
#New Shape of DataFrame
df.shape
```

### Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
```

```python
#Normal Histogram plot for Points column
fig=plt.figure()
ax=fig.add_axes([0,0,4,2])
ax.hist(df.PTS)
```

```python
#Modified Histogram plot for points column
fig=plt.figure()
ax=fig.add_axes([0,0,4,2])
ax.set_title('Points Table through Histogram')
ax.set_xlabel('Points on the Table')
ax.hist(df.PTS,bins=[0,5,10,15,20,25,30,35],rwidth=0.8)
```
